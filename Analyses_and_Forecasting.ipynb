```python
dataset1 = pd.read_csv('/content/drive/MyDrive/FYP - DATASET/taxi+_zone_lookup.csv')
dataset2 = pd.read_csv('/content/drive/MyDrive/FYP - DATASET/11.csv')

dataset1.info()
dataset2.info()

# Sort dataset1 in ascending order based on LocationID
dataset1_sorted = dataset1.sort_values(by='LocationID')

# Display the sorted dataset
print(dataset1_sorted)
```

```python

# Sort dataset1 in ascending order based on LocationID
dataset2_sorted = dataset2.sort_values(by='locationID')

# Display the sorted dataset
print(dataset2_sorted)
```

```python
# Remove rows with locationID 264 and 265
dataset2_filtered = dataset2_sorted[~dataset2_sorted['locationID'].isin([264, 265])]

# Display the filtered dataset
print(dataset2_filtered)
```

```python
# Merge dataset2_filtered with dataset1_sorted based on locationID
merged_dataset = pd.merge(dataset2_filtered, dataset1_sorted, how='left', left_on='locationID', right_on='LocationID', suffixes=('_dataset2', '_dataset1'))

# Display the merged dataset
print(merged_dataset)

merged_dataset.head(1000)
```

```python
# Check for missing values
missing_values = merged_dataset.isnull().sum()
print(missing_values)
```

```python
# Assuming your DataFrame is named merged_dataset
merged_dataset = merged_dataset.dropna(subset=['Affiliated_base_num'])

# Display the updated DataFrame
print(merged_dataset)

missing_values = merged_dataset.isnull().sum()
print(missing_values)
```

### **Outliers detection and removal using Z-score:**

```python
from scipy.stats import zscore

# Assuming your DataFrame is named 'merged_dataset'
numeric_columns = ['locationID', 'LocationID']  # Add other numeric columns as needed

# Calculate Z-scores for each numeric column
z_scores = zscore(merged_dataset[numeric_columns])

# Identify and remove outliers for each numeric column
outliers = (abs(z_scores) > 3)
merged_dataset_no_outliers = merged_dataset[~outliers.all(axis=1)]

# Display the updated DataFrame without outliers
print(merged_dataset_no_outliers)
```

```python
merged_dataset.head()
```

### **Identifying Boroughs**

```python
# Assuming 'merged_dataset' is your DataFrame
boroughs = merged_dataset['Borough'].unique()

# Split the dataset into separate DataFrames for each borough
borough_data = {}
for borough in boroughs:
    borough_data[borough] = merged_dataset[merged_dataset['Borough'] == borough]

# Display DataFrames for each borough
for borough, data in borough_data.items():
    print(f'\nData for Borough: {borough}')
    print(data)


```

### **Normalize Dataset using Min-Max Scaling**

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming 'numeric_columns' are the columns you want to normalize
numeric_columns = ['locationID', 'LocationID']  # Add other numeric columns as needed

# Initialize Min-Max scaler
scaler = MinMaxScaler()

# Normalize the numeric columns for each borough
for borough, data in borough_data.items():
    data[numeric_columns] = scaler.fit_transform(data[numeric_columns])

# Display the normalized DataFrame for each borough
print(f'\nNormalized Data for Borough: {borough}')
print(data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from matplotlib.dates import DateFormatter

# Make sure 'Pickup_date' is in datetime format
copied_dataset['Pickup_date'] = pd.to_datetime(copied_dataset['Pickup_date'])

# Create 'noOfRides' by counting the number of rides for each 'Pickup_date'
noOfRides = copied_dataset.groupby('Pickup_date').size().reset_index(name='noOfRides')

# Now 'noOfRides' DataFrame has two columns: 'Pickup_date' and 'noOfRides'
# Set 'Pickup_date' as the index for the noOfRides DataFrame
noOfRides.set_index('Pickup_date', inplace=True)

# ARIMA model fitting, the order (p,d,q) needs to be determined by analysis
# For demonstration, we use order=(1, 1, 1)
model = ARIMA(noOfRides['noOfRides'], order=(1, 1, 1))
model_fit = model.fit()

# Forecast
noOfRides['ARIMA_Forecasting'] = model_fit.predict(start=0, end=len(noOfRides) - 1, typ='levels')

# Plotting the original data and the forecasting
plt.figure(figsize=(10, 5))
plt.plot(noOfRides.index, noOfRides['noOfRides'], label='Original')
plt.plot(noOfRides.index, noOfRides['ARIMA_Forecasting'], label='ARIMA Forecasting', color='orange')
plt.legend()

# Formatting the plot
plt.title('Forecasting number of rides using ARIMA model')
plt.xlabel('Pickup date')
plt.ylabel('Number of Rides')
plt.gca().xaxis.set_major_formatter(DateFormatter('%d %b'))
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from matplotlib.dates import DateFormatter

# Make sure 'Pickup_date' is in datetime format
copied_dataset['Pickup_date'] = pd.to_datetime(copied_dataset['Pickup_date'])

# Create 'noOfRides' by counting the number of rides for each 'Pickup_date'
noOfRides = copied_dataset.groupby('Pickup_date').size().reset_index(name='noOfRides')

# ARIMA model fitting, the order (p,d,q) needs to be determined by analysis
# For demonstration, we use order=(1, 1, 1)
model = ARIMA(noOfRides['noOfRides'], order=(1, 1, 1))
model_fit = model.fit()

# Create a date range for forecasting
forecast_start = noOfRides['Pickup_date'].max() + pd.DateOffset(1)
forecast_end = noOfRides['Pickup_date'].max() + pd.DateOffset(len(noOfRides))
forecast_index = pd.date_range(start=forecast_start, end=forecast_end, freq='D')

# Forecast
forecast = model_fit.predict(start=len(noOfRides), end=len(noOfRides) + len(forecast_index) - 1, typ='levels')

# Create a DataFrame for the forecast with the correct index
forecast_df = pd.DataFrame(forecast, index=forecast_index, columns=['ARIMA_Forecast'])

# Plotting the ARIMA forecast with an orange line
plt.figure(figsize=(15, 5))
plt.plot(noOfRides['Pickup_date'], noOfRides['noOfRides'], label='Daily Uber Rides', color='blue')
plt.plot(forecast_df.index, forecast_df['ARIMA_Forecast'], label='ARIMA Forecast', color='orange')
plt.xlabel('Pickup Date')
plt.ylabel('Number of Rides')
plt.title('Daily Uber Rides and ARIMA Forecast')
plt.gca().xaxis.set_major_formatter(DateFormatter('%d %b'))
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot with ARIMA forecasting and legend
plt.legend()
plt.show()
```

```python
# Assuming 'bronx_data' is already filtered for the Bronx borough
# Group by 'Pickup_date' to get the number of rides per day
daily_rides = bronx_data.resample('D').size()

# Define the size of the test set
TEST_SIZE = 30  # For example, last 30 days of the dataset

# Split the data
train = daily_rides.iloc[:-TEST_SIZE]
test = daily_rides.iloc[-TEST_SIZE:]

from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model
# Here, I'm using (1, 1, 1) as example parameters
model = ARIMA(train, order=(1, 1, 1))
model_fit = model.fit()

# Forecast the values for the test set duration
forecast = model_fit.get_forecast(steps=TEST_SIZE)
predicted_mean = forecast.predicted_mean

fig, ax = plt.subplots(figsize=(15, 5))

# Plot training data
ax.plot(train.index, train, label='Training Data')

# Plot test data
ax.plot(test.index, test, label='Test Data', color='orange')

# Plot predicted data
ax.plot(predicted_mean.index, predicted_mean, label='Forecast', color='green')

# Add labels and title
plt.xlabel('Pickup Date')
plt.ylabel('Number of Rides')
plt.title('Daily Uber Rides in Bronx with ARIMA Forecasting')
plt.legend()
plt.show()
```
