# Forecasting number of rides using ARIMA model

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from matplotlib.dates import DateFormatter

# Make sure 'Pickup_date' is in datetime format
copied_dataset['Pickup_date'] = pd.to_datetime(copied_dataset['Pickup_date'])

# Create 'noOfRides' by counting the number of rides for each 'Pickup_date'
noOfRides = copied_dataset.groupby('Pickup_date').size().reset_index(name='noOfRides')

# Now 'noOfRides' DataFrame has two columns: 'Pickup_date' and 'noOfRides'
# Set 'Pickup_date' as the index for the noOfRides DataFrame
noOfRides.set_index('Pickup_date', inplace=True)

# ARIMA model fitting, the order (p,d,q) needs to be determined by analysis
# For demonstration, we use order=(1, 1, 1)
model = ARIMA(noOfRides['noOfRides'], order=(1, 1, 1))
model_fit = model.fit()

# Forecast
noOfRides['ARIMA_Forecasting'] = model_fit.predict(start=0, end=len(noOfRides) - 1, typ='levels')

# Plotting the original data and the forecasting
plt.figure(figsize=(10, 5))
plt.plot(noOfRides.index, noOfRides['noOfRides'], label='Original')
plt.plot(noOfRides.index, noOfRides['ARIMA_Forecasting'], label='ARIMA Forecasting', color='orange')
plt.legend()

# Formatting the plot
plt.title('Forecasting number of rides using ARIMA model')
plt.xlabel('Pickup date')
plt.ylabel('Number of Rides')
plt.gca().xaxis.set_major_formatter(DateFormatter('%d %b'))
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Assuming 'bronx_data' is already filtered for the Bronx borough
# Group by 'Pickup_date' to get the number of rides per day
daily_rides = bronx_data.resample('D').size()

# Define the size of the test set
TEST_SIZE = 30  # For example, last 30 days of the dataset

# Split the data
train = daily_rides.iloc[:-TEST_SIZE]
test = daily_rides.iloc[-TEST_SIZE:]

from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model
# Here, I'm using (1, 1, 1) as example parameters
model = ARIMA(train, order=(10, 1, 1))
model_fit = model.fit()

# Forecast the values for the test set duration
forecast = model_fit.get_forecast(steps=TEST_SIZE)
predicted_mean = forecast.predicted_mean

fig, ax = plt.subplots(figsize=(15, 5))

# Plot training data
ax.plot(train.index, train, label='Training Data')

# Plot test data
ax.plot(test.index, test, label='Test Data', color='orange')

# Plot predicted data
ax.plot(predicted_mean.index, predicted_mean, label='Forecast', color='green')

# Add labels and title
plt.xlabel('Pickup Date')
plt.ylabel('Number of Rides')
plt.title('Daily Uber Rides in Bronx with ARIMA Forecasting')
plt.legend()
plt.show()
```

## Using pickup_counts_ld

```python
import numpy as np

def pickup_counts_ld_to_x_y(pickup_counts_ld, window_size):
    # Convert 'Count' column to a numpy array
    counts = pickup_counts_ld['Count'].to_numpy()
    X = []
    Y = []
    for i in range(len(counts) - window_size):
      window = counts[i:i + window_size]
      X.append(window)
      label = counts[i + window_size]
      Y.append(label)
    return np.array(X), np.array(Y)
```

```python
# Example usage
window_size = 3  # You can adjust this to your desired window size
X, Y = pickup_counts_ld_to_x_y(pickup_counts_ld, window_size)
X.shape , Y.shape
```

```python
X_train, Y_train =X[:90],Y[:90]

X_val, Y_val = X[90:95],Y[90:95]

X_test, Y_test =X[95:],Y[95:]

X_train.shape, Y_train.shape, X_val.shape, Y_val.shape, X_test.shape, Y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1 = Sequential()
model1.add(InputLayer((3,1)))
model1.add(LSTM(64))
model1.add(Dense(8,'relu'))
model1.add(Dense(1,'linear'))

model1.summary()
```

```python
cp = ModelCheckpoint('model1/',save_best_only=True)
model1.compile(loss=MeanSquaredError(),optimizer=Adam(learning_rate=0.0001),metrics=[RootMeanSquaredError()])
```

```python
model1.fit(X_train, Y_train, validation_data=(X_val,Y_val),epochs=10, callbacks=[cp])
```
