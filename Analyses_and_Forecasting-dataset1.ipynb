### **Outliers detection and removal using Z-score:**

```python
from scipy.stats import zscore

# Assuming your DataFrame is named 'merged_dataset'
numeric_columns = ['locationID', 'LocationID']  # Add other numeric columns as needed

# Calculate Z-scores for each numeric column
z_scores = zscore(merged_dataset[numeric_columns])

# Identify and remove outliers for each numeric column
outliers = (abs(z_scores) > 3)
merged_dataset_no_outliers = merged_dataset[~outliers.all(axis=1)]

# Display the updated DataFrame without outliers
print(merged_dataset_no_outliers)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from matplotlib.dates import DateFormatter

# Make sure 'Pickup_date' is in datetime format
copied_dataset['Pickup_date'] = pd.to_datetime(copied_dataset['Pickup_date'])

# Create 'noOfRides' by counting the number of rides for each 'Pickup_date'
noOfRides = copied_dataset.groupby('Pickup_date').size().reset_index(name='noOfRides')

# ARIMA model fitting, the order (p,d,q) needs to be determined by analysis
# For demonstration, we use order=(1, 1, 1)
model = ARIMA(noOfRides['noOfRides'], order=(1, 1, 1))
model_fit = model.fit()

# Create a date range for forecasting
forecast_start = noOfRides['Pickup_date'].max() + pd.DateOffset(1)
forecast_end = noOfRides['Pickup_date'].max() + pd.DateOffset(len(noOfRides))
forecast_index = pd.date_range(start=forecast_start, end=forecast_end, freq='D')

# Forecast
forecast = model_fit.predict(start=len(noOfRides), end=len(noOfRides) + len(forecast_index) - 1, typ='levels')

# Create a DataFrame for the forecast with the correct index
forecast_df = pd.DataFrame(forecast, index=forecast_index, columns=['ARIMA_Forecast'])

# Plot the daily rides
plt.figure(figsize=(15, 5))
plt.plot(noOfRides['Pickup_date'], noOfRides['noOfRides'], label='Daily Uber Rides', color='blue')
plt.xlabel('Pickup Date')
plt.ylabel('Number of Rides')

# Plotting the ARIMA forecast with an orange line
plt.plot(forecast_df.index, forecast_df['ARIMA_Forecast'], label='ARIMA Forecast', color='orange')

plt.title('Daily Uber Rides and ARIMA Forecast')
plt.gca().xaxis.set_major_formatter(DateFormatter('%d %b'))
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot with ARIMA forecasting and legend
plt.legend()
plt.show()
```
